```python
import os
import urllib3
import tabula
import pandas as pd


urllib3.disable_warnings()
url = "https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3164749/pdf/nihms303675.pdf"

fileName = r"file.pdf"
with urllib3.PoolManager() as http:
    r = http.request('GET', url)
    with open(fileName, 'wb') as fout:
        fout.write(r.data)


cwd=os.getcwd()
path = cwd  +"/"+ fileName
df = tabula.read_pdf(path, pages = 'all', multiple_tables = True)
print(df)
#type(df)
#df2=df
#df2.to_csv('columns.csv')
```
